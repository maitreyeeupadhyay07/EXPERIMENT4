```python
#set operation
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("UNION:",A | B)
print("INTERSECTION:",A & B)
print("DIFFERENCE:",A - B)
print("DIFFERENCE2:", B - A)
print("SYMMETRIC DIFFERENCE:",A ^ B)
```

```python
# PROBELEM STATEMENT 1: UNIQUE EVENT PARTICIPANT
#Q] A COLLEGE EVENT HAS A LIST OF STUDENTS WHO REGISTERED MULTIPLE TIME BY MISTAKE. WRITE A PROGRAM USING SET TO STORE PARTIICIPANT NAMES AND DISPLAY ONLY UNIQUE PARTCIPANTS
studentlist = ["riya","niya","tiya","piya","jiya","riya"]
unique_studentlist =set(studentlist)
print(" UNIQUE PARTICIPANTS:" , unique_studentlist)
```
